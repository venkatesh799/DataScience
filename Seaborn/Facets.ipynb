```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel("Sales.xlsx")
```

```python
data.head()
```

###### The best approach may be to make more than one plot. Because relplot() is based on the FacetGrid, this is easy to do.

###### To show the influence of an additional variable, instead of assigning it to one of the semantic roles in the plot, use it to 

###### “facet” the visualization.

######  This means that you make multiple axes and plot subsets of the data on each of them:

```python
sns.relplot("Sales","Profit",data=data)
```

```python
sns.relplot("Sales","Profit",data=data,col="Customer Segment")
```

```python
sns.relplot("Sales","Profit",data=data,col="Customer Segment",hue="Ship Mode",col_wrap=2)
```

```python
sns.relplot("Sales","Profit",data=data,col="Customer Segment",hue="Ship Mode",row="Product Category")
```
