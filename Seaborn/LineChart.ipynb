# linePlot

#### With some datasets, you may want to understand changes in one variable as a function of time, or a similarly continuous variable. In this situation, a good choice is to draw a line plot. In seaborn, this can be accomplished by the lineplot() function, either directly or with relplot() by setting kind="line":

```python
sns.relplot(x="Order Date",y="Sales",data=data,kind="line")
```

```python
data["year"] = data["Order Date"].dt.year
```

```python
sns.relplot(x="year",y="Profit",data=data,kind="line")
```

#### Actually it calculated the mean and The Shaded area is called +95%  and   -95%  confidence interval of the mean

#### To remove this ci(ConfidenceInterval) = None

```python
sns.relplot("year","Profit",kind="line",data=data,ci=None)
```

#### Lineplot Automatically sorts the data and provides the plot for us.we can disable it 

#### sort = False

```python
sns.relplot("year","Profit",kind="line",data=data,sort=False,ci=None)
```

#### Another good option, especially with larger data, is to represent the spread of the distribution at each timepoint by plotting the standard deviation instead of a confidence interval:

```python
sns.relplot("year","Profit",kind="line",data=data,ci="sd")
```

```python
sns.relplot("year","Profit",kind="line",data=data,estimator=None)
```

```python
sns.relplot(x="year",y="Sales",kind="line",data=data,hue="Ship Mode")
```

```python
sns.relplot(x="year",y="Sales",kind="line",data=data,hue="Ship Mode",style="Product Category")
```

```python
sns.relplot(x="year",y="Sales",kind="line",data=data,hue="Ship Mode",style="Product Category",ci=None)
```

```python
sns.relplot(x="year",y="Sales",kind="line",data=data,hue="Ship Mode",col="Product Category",ci=None)
```

```python
sns.relplot(x="year",y="Sales",kind="line",data=data,hue="Ship Mode",col="Region",ci=None)
```

```python
sns.relplot(x="year",y="Sales",kind="line",data=data,hue="Ship Mode",col="Region",ci=None,col_wrap=4)
```
