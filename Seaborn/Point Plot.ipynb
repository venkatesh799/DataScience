```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel("sales.xlsx")
```

```python
data.head()
```

```python
sns.catplot(x="Product Category",y="Sales",data=data,kind="point")
```

```python
sns.catplot(x="Product Category",y="Sales",data=data,kind="point",hue="Ship Mode",linestyles=["-","--","-."])
```

```python
sns.catplot(x="Product Category",y="Sales",data=data,kind="point",hue="Ship Mode",col="Customer Segment")
```
