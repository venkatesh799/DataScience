```python
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_excel("Sales.xlsx")
```

```python
data.head()
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",kind="box",data=data)
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",kind="box",hue="Ship Mode",data=data)
```

#### i want the above chart in different charts for different ship mode

```python
sns.catplot(x="Product Category",y="Shipping Cost",kind="box",row="Ship Mode",data=data)
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",kind="box",col="Ship Mode",data=data)
```

#### A related function, boxenplot(), draws a plot that is similar to a box plot but optimized for showing more information about the shape of the distribution. It is best suited for larger datasets:

```python
sns.catplot(x="Product Category",y="Shipping Cost",kind="boxen",data=data)
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",kind="boxen",col="Customer Segment",row="Ship Mode",data=data)
```
