```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel("sales.xlsx")
```

```python
data.head()
```

```python
sns.relplot(x="Sales",y="Profit",data=data)
```

### It’s also possible to represent four variables by changing the hue and style of each point independently. But this should be done carefully, because the eye is much less sensitive to shape than to color:

```python
sns.relplot(x="Sales",y="Profit",hue="Order Priority",data=data)
```

### It’s also possible to represent four variables by changing the hue and style of each point independently. But this should be done carefully, because the eye is much less sensitive to shape than to color

```python
sns.relplot(x="Sales",y="Profit",style="Order Priority",hue="Order Priority",data=data)
```

```python
sns.relplot(x="Sales",y="Profit",hue="Order Priority",style="Ship Mode",data=data)
```

```python
sns.relplot(x="Sales",y="Profit",size="Discount",data=data)
```

```python
sns.relplot(x="Sales",y="Profit",size="Discount",sizes=(25,200),data=data)
```

```python
sns.relplot(x="Sales",y="Profit",hue="Discount",data=data)
```
